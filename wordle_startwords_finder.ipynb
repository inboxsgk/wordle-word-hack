## Finding frequency of letters in the dataset

```python
c = []
occurance_letter = {}
for i in dataset:
  cur = []
  for j in i:
    cur.append(j)
  #print(cur)
  for j in i:
    if j not in c:
      c.append(j)
      occurance_letter[j] = 1
    elif j in c:
      if cur.count(j)==1:
        occurance_letter[j] += 1
```

```python
occurance_letter = dict(sorted(occurance_letter.items(), key=lambda item: item[1]))
t=0
for i in occurance_letter:
  print(i, "-", occurance_letter[i])
  t+=occurance_letter[i]
  
print(t, "occurances")
print(len(occurance_letter), "letters")
```

## Calculating probability of select character present in any word

```python
pr = {}
for i in occurance_letter:
  pr[i] = (occurance_letter[i]/t)
  print(i, " - ", pr[i])
```

## Generating Probability for every word in dataset based on letters frequency

```python
scores={}
for i in dataset:
  cur = 0
  for j in i:
    cur += pr[j]
  scores[i]=cur

scores = dict(sorted(scores.items(), key=lambda item: item[1]))
```

```python
l_final = []
for i in scores:
  print(i," - ", scores[i])
  l_final.append(i)
```

```python
l_final = l_final[::-1]
```

## Finding the best startword
Searching for words with unique characters (Non-repetitive) to stategically exhaust as many characters we can in the first few tries

```python
letters_used = []
words_found = []
print("Done")
```

```python
for i in l_final:
  b = 0
  c = []
  for j in i:
    if j not in c:
      c.append(j)
    elif j in c:
      b=1
      break
  if (b==1):
    continue
  for j in i:
    if j in letters_used:
      b+=1
      break
  if (b==1):
    continue
  else :
    print("Found: ", i)
    words_found.append(i)
    for j in i:
      letters_used.append(j)
```

## Printing the results

```python
print(len(words_found), "words found!")
print(words_found)
print(letters_used)
```
